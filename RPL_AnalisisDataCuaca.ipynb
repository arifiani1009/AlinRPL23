```python
#rawdata berisi tabel dengan kolom yang diuraikan
rawdata = df[['dt_iso', 'city_name', 'temp', 'pressure', 'humidity', 'wind_speed', 'wind_deg', 'weather_id', 'weather_main']]
rawdata
```

```python
#tabel berisi kolom humidity dan temp
weather_temp = rawdata[["humidity","temp"]]
weather_temp.head(10)
```

```python
dummies = pd.get_dummies(rawdata["weather_main"])
dummies.head(122)
```

```python
weather_temp2 = pd.concat([weather_temp,dummies],axis=1)
weather_temp2.head(10)
```

```python
#  Spliting X (Independent Variables) and Y(predicting variable)
Y = weather_temp["temp"]
X = weather_temp2
# making Two dimensional array
X
```

```python
# Normalizing the X variables 
# check the x values should be between (-3  3)
#X.dropna()
x = (X >=0)
x.sum()
# prints the zero humidity
print(x.sum(),X.shape)
non_zero_humidity = X >= 0
print(non_zero_humidity.sum())
```

```python
#split data test and data train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

```python
X_train.shape, y_train.shape
```

```python
#print(X_train.head(10))
y_train.head(10)
```

```python
#data training dg linier regresssion
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_train.shape,y_train.shape
```

```python
plt.scatter(X_train["humidity"], y_train)
#plt.plot(X_train["humidity"], model.predict(X_train["humidity"]), color='black')
plt.title("Temperature v/s humidity")
plt.xlabel("Humidity")
plt.ylabel("temperature")
```

```python
#dengan data test
pred = model.predict(X_test)
pred
```
